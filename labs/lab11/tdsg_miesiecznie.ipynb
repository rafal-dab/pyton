```python
import pandas as pd
import numpy as np
```

```python
tdsg = pd.read_csv('TDSG_024_2021_60min_nospec.csv', index_col=0, parse_dates=True, dayfirst=True)
```

```python
#tdsg = pd.read_csv('TDSG_024_2021_60min_nospec.csv', index_col=0, parse_dates=True)    
h10 = tdsg[['GMfromSpectrum', 'MeteoRainfall']].fillna(0)

print("Liczba wszystkich wyników: ", h10['GMfromSpectrum'].count())
deszcz_idx = h10[h10['MeteoRainfall'] > 0].index
print("Liczba wyników z deszczem: ", len(deszcz_idx))
print("Liczba wyników bez deszczu: ", h10['GMfromSpectrum'].count() - len(deszcz_idx))

for i in deszcz_idx:
    #print(i)
    idx = pd.date_range(i - pd.Timedelta("1H"), i + pd.Timedelta("2H"), freq='H')
    #print(idx)
    #break
    h10.drop(idx, inplace=True, errors='ignore')
#h10.drop(idx, inplace=True, errors='ignore')

print("Liczba wyników do obliczeń: ", h10['GMfromSpectrum'].count())
#print(h10)

h10_mon = h10["GMfromSpectrum"].resample('M') #, closed='left')
#print(h10_mon)
h10_mon.agg([np.size, np.min, np.max, np.mean, np.std])
#h10_mon.sum()
```
